```python
documents = ["AI is transforming our world",
"Automating tasks with precision and speed",
"Machine learning algorithms learn and grow",
"Revolutionizing industries, high and low."]
#query = input("Enter your query: ")
query = "Who is transforming our world"
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
HuggingFaceEmbeddings_ = HuggingFaceEmbeddings(model_name='all-MiniLM-L6-v2')

# to embed a list of the text
documents_embed = HuggingFaceEmbeddings_.embed_documents(documents)

query_embed = HuggingFaceEmbeddings_.embed_query(query)
```

```python
def similarity_search (document_embeddings, query_embeddings) -> int:
    # similarity search based on the cosine similaarity
    from numpy import dot, linalg
    dot_product = dot(document_embeddings, query_embeddings)
    norm1 = linalg.norm(document_embeddings)
    norm2 = linalg.norm(query_embeddings)
    return dot_product/(norm1*norm2)

most_related_ranking = []
for i in range(len(documents)):
    # prints the documents with their correspoinding embeddings
    print(documents[i], documents_embed[i])
    value = similarity_search(documents_embed[i], query_embed)
    print(i, " ->>>> ", value)
    most_related_ranking.append(value)

for i in range(len(most_related_ranking)):
    print(i, most_related_ranking[i])
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
from langchain.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from typing import List, Dict, Any
import os
import tiktoken
import logging
from dataclasses import dataclass
from datetime import datetime

# Configure logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

@dataclass
class DocumentMetadata:
    source: str
    chunk_index: int
    total_chunks: int
    processing_date: str

class DocumentProcessor:
    def __init__(self, model_name: str = 'gpt-4', chunk_size: int = 1000, chunk_overlap: int = 200):
        self.chunk_size = chunk_size
        self.chunk_overlap = chunk_overlap
        self.model_name = model_name
        self.setup_components()


    def setup_components(self):
        """Initialize all necessary components with improved configuration"""
        self.text_splitter = RecursiveCharacterTextSplitter(
            chunk_size=self.chunk_size,
            chunk_overlap=self.chunk_overlap,
            length_function=len,
            separators=["\n\n", "\n", ".", " ", ""],
            add_start_index=True,
        )

        self.embedding_model = HuggingFaceEmbeddings(
            model_name='all-MiniLM-L6-v2',
            model_kwargs={'device': 'cpu'},
            encode_kwargs={'normalize_embeddings': True}
        )

        self.tokenizer = tiktoken.get_encoding("cl100k_base")

    def load_and_process_document(self, file_path: str) -> List[Dict[str, Any]]:
        """Load and process document with enhanced metadata"""
        try:
            # Primero verificamos si el archivo existe
            if not os.path.exists(file_path):
                raise FileNotFoundError(f"El archivo {file_path} no existe")

            # Leemos el contenido del archivo directamente para verificar
            with open(file_path, 'r', encoding='utf-8') as file:
                content = file.read()
                logger.info(f"Contenido leído: primeros 100 caracteres: {content[:100]}")

            loader = TextLoader(file_path, encoding='utf-8')
            documents = loader.load()

            # Logging para verificar el contenido cargado
            logger.info(f"Número de documentos cargados: {len(documents)}")
            if documents:
                logger.info(f"Muestra del primer documento: {documents[0].page_content[:100]}")

            # Enhanced splitting with metadata
            splits = self.text_splitter.split_documents(documents)
            logger.info(f"Número de splits generados: {len(splits)}")

            processed_docs = []
            for i, doc in enumerate(splits):
                metadata = {
                    'source': file_path,
                    'chunk_index': i,
                    'total_chunks': len(splits),
                    'processing_date': datetime.now().isoformat(),
                    'token_count': len(self.tokenizer.encode(doc.page_content))
                }
                doc.metadata = metadata
                processed_docs.append(doc)

            return processed_docs
        except Exception as e:
            logger.error(f"Error processing document: {str(e)}")
            raise

    def create_vector_store(self, documents: List[Dict[str, Any]]) -> Chroma:
        """Create and configure vector store with improved retrieval settings"""
        try:
            return Chroma.from_documents(
                documents=documents,
                embedding=self.embedding_model,
                collection_metadata={"hnsw:space": "cosine"}
            )
        except Exception as e:
            logger.error(f"Error creating vector store: {str(e)}")
            raise
    def get_relevant_context(self, query: str, vector_store: Chroma, max_tokens: int = 6000) -> str:
        """Get relevant context with improved filtering and processing"""
        try:
            retriever = vector_store.as_retriever(
                search_type="similarity",
                search_kwargs={
                    "k": 4,
                    "filter": None  # Aseguramos que no haya filtros que limiten la búsqueda
                }
            )

            relevant_docs = retriever.get_relevant_documents(query)
            logger.info(f"Número de documentos relevantes encontrados: {len(relevant_docs)}")

            processed_contexts = []
            current_tokens = 0

            for doc in relevant_docs:
                # Log para debug
                logger.info(f"Procesando documento con contenido: {doc.page_content[:100]}...")

                chunk_tokens = len(self.tokenizer.encode(doc.page_content))

                if current_tokens + chunk_tokens > max_tokens:
                    break

                processed_contexts.append(f"{doc.page_content}")
                current_tokens += chunk_tokens

            final_context = "\n\n".join(processed_contexts)
            logger.info(f"Longitud del contexto final: {len(final_context)} caracteres")

            return final_context if processed_contexts else "No se encontró contexto relevante."

        except Exception as e:
            logger.error(f"Error getting relevant context: {str(e)}")
            raise

    def create_prompt_template(self) -> ChatPromptTemplate:
        """Create an improved prompt template with better guidance"""
        template = """
        Basándote en el siguiente contexto del documento, proporciona un análisis detallado y profundo.

        Contexto:
        {context}

        Pregunta: {question}

        Instrucciones específicas:
        1. Analiza el texto proporcionado
        2. Identifica los elementos clave
        3. Proporciona citas textuales relevantes para respaldar el análisis
        4. Explica las implicaciones principales
        5. Considera diferentes interpretaciones cuando sea pertinente

        Por favor, proporciona un análisis claro y estructurado.
        """
        return ChatPromptTemplate.from_template(template)

class QueryProcessor:
    def __init__(self, doc_processor: DocumentProcessor):
        self.doc_processor = doc_processor
        self.llm = ChatOpenAI(
            model=doc_processor.model_name,
            temperature=0.3,
            presence_penalty=0.3,
            frequency_penalty=0.3,
        )

    def process_query(self, query: str, vector_store: Chroma) -> str:
        """Process query with improved error handling and response quality"""
        try:
            # Get context
            context = self.doc_processor.get_relevant_context(query, vector_store)
            if context.startswith("Error"):
                return context

            # Create and execute chain
            prompt = self.doc_processor.create_prompt_template()
            chain = prompt | self.llm

            # Get response
            response = chain.invoke({
                "context": context,
                "question": query
            })

            return response.content

        except Exception as e:
            logger.error(f"Error processing query: {str(e)}")
            return f"Error al procesar la consulta: {str(e)}"

def main():
    # Initialize processors
    doc_processor = DocumentProcessor()
    query_processor = QueryProcessor(doc_processor)

    try:
        # Definir la ruta del archivo
        file_path = '/content/evangelio_segun_marcos.txt'

        # Verificar que el archivo existe
        logger.info(f"Verificando archivo en: {file_path}")
        if not os.path.exists(file_path):
            raise FileNotFoundError(f"El archivo {file_path} no existe")

        # Verificar que podemos leer el archivo
        with open(file_path, 'r', encoding='utf-8') as f:
            sample = f.read(100)
            logger.info(f"Muestra del contenido del archivo: {sample}")

        # Process document
        logger.info("Cargando y procesando el documento...")
        documents = doc_processor.load_and_process_document(file_path)
        logger.info(f"Documentos procesados: {len(documents)}")

        logger.info("Creando vector store...")
        vector_store = doc_processor.create_vector_store(documents)

        while True:
            # Get user query
            query = input("\nIngrese su consulta sobre el texto (o 'salir' para terminar): ")

            if query.lower() == 'salir':
                break

            logger.info("Procesando consulta...")
            # Process query and get response
            response = query_processor.process_query(query, vector_store)

            # Print response
            print("\nRespuesta:")
            print("="*80)
            print(response)
            print("="*80)

    except Exception as e:
        logger.error(f"Error en la ejecución: {str(e)}")
        print(f"\nError: {str(e)}")

if __name__ == "__main__":
    main()
```
